### 1st step is to drop duplicates

```python
backPackData.drop_duplicates(subset=["product_id"], inplace=True)
```

### I saw that most of the values in what_customers_said column are missing, so we are droping it :)

```python
backPackData = backPackData.drop('what_customers_said', axis=1)
```

### I also saw that seller_information is also null Column so let's drop it

```python
backPackData = backPackData.drop('seller_information', axis=1)
```

```python
print(backPackData.to_string())
```

### Let's Filter the data


### where rating is greater than 4

```python
highRating = backPackData[backPackData["rating"] > 4]
print(highRating.head())
```

### Selecting title, rating, finalPrice where rating is greater than 4.5

```python
result = backPackData.loc[
    backPackData["rating"] >= 4.5,
    ["title", "rating", "final_price"]
]
print(result.head())
```

### Selecting only what i want to see

```python
selected = backPackData[[
    "product_id",
    "title",
    "rating",
    "final_price",
    "category"
]]

print(selected.head())
```
